## 1.1 Moore’s Law

* **Growth Factor Required:** 1,000 / 1.809 ≈ 552.8x

## 1.2 Amdahl’s Law

### 1.2.1 Performance Improvement (Amdahl's Law)

**Formula:**
$$a = \frac{1}{(1 - p) + \frac{p}{n}}$$

**Given Variables:**
* **$p$ (Parallel/Improved fraction):** 40% or **0.4** (Time spent on CPU calculations)
* **$1 - p$ (Serial/Unimproved fraction):** 60% or **0.6** (Time spent on I/O)
* **$n$ (Execution units / Speed-up factor):** **10** (The new CPU is 10x faster)

**Calculation:**
$$a = \frac{1}{(1 - 0.4) + \frac{0.4}{10}}$$
$$a = \frac{1}{0.6 + 0.04}$$
$$a = \frac{1}{0.64}$$
$$a = 1.5625$$

**Result:**
The expected overall performance improvement is **1.5625 times** (a 56.25% increase).

### 1.2.2 Optimization Comparison (Amdahl's Law)

**Formula:**
$$a = \frac{1}{(1 - p) + \frac{p}{n}}$$

#### Alternative 1: Improve FPSQR only
* **$p$ (FPSQR fraction):** 20% or **0.2**
* **$n$ (Speed-up factor):** **10**
$$a_1 = \frac{1}{(1 - 0.2) + \frac{0.2}{10}} = \frac{1}{0.8 + 0.02} = \frac{1}{0.82} \approx 1.2195$$
* **Speedup:** ~1.22x

#### Alternative 2: Improve all FP operations
* **$p$ (FP operations fraction):** 50% or **0.5** (since half of the execution time is spent on FP)
* **$n$ (Speed-up factor):** **1.6**
$$a_2 = \frac{1}{(1 - 0.5) + \frac{0.5}{1.6}} = \frac{1}{0.5 + 0.3125} = \frac{1}{0.8125} \approx 1.2308$$
* **Speedup:** ~1.23x

### Conclusion
**Alternative 2 is the optimal solution.** Even though the individual speed-up factor is much smaller (1.6x vs 10x), it applies to a significantly larger fraction of the overall execution time (50% vs 20%), resulting in a higher total system speedup (1.23x vs 1.22x).

### 1.2.3 Maximum Serial Fraction (Amdahl's Law)

**Formula:**
$$a = \frac{1}{(1 - p) + \frac{p}{n}}$$

Where $(1 - p)$ is the serial fraction. Let's substitute $(1 - p)$ with $s$ (so $p = 1 - s$) for easier solving:
$$a = \frac{1}{s + \frac{1 - s}{n}}$$

**Given Variables:**
* **$a$ (Target speedup):** 100
* **$n$ (Number of processors):** 128
* **$s$ (Serial fraction $1-p$):** Unknown

**Calculation:**
$$100 = \frac{1}{s + \frac{1 - s}{128}}$$
$$100 \cdot \left(s + \frac{1 - s}{128}\right) = 1$$
$$100s + \frac{100 - 100s}{128} = 1$$

Multiply the entire equation by 128 to remove the fraction:
$$12800s + 100 - 100s = 128$$
$$12700s = 28$$
$$s = \frac{28}{12700} \approx 0.0022047$$

**Result:**
To achieve a 100x speedup on 128 processors, the serial fraction ($1-p$) can be at most **0.22%**.
